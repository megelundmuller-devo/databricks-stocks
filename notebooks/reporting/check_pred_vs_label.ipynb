```sql
CREATE OR REPLACE TEMP VIEW stocks_reporting AS
SELECT p.*, abs(f.feat_intraday_return) as percent_change, case when p.prediction = p.label then 1 else 0 end as correct FROM stocks.reporting.predictions p
LEFT JOIN stocks.silver.stocks_w_features f ON
  p.date = f.date 
  and p.company = f.company
```

```sql
select * from stocks_reporting 
where certainty not between 0.1 and 0.9
order by date desc
```

```sql
select '10%' as interval, sum(correct)/sum(CASE WHEN label is not null then 1 END) as correct, sum(CASE WHEN label is not null then 1 END) as entries from stocks_reporting 
where certainty not between 0.1 and 0.9
UNION
select '20%' as interval, sum(correct)/sum(CASE WHEN label is not null then 1 END) as correct, sum(CASE WHEN label is not null then 1 END) as entries from stocks_reporting 
where certainty not between 0.2 and 0.8
union
select '30%' as interval, sum(correct)/sum(CASE WHEN label is not null then 1 END) as correct, sum(CASE WHEN label is not null then 1 END) as entries from stocks_reporting 
where certainty not between 0.3 and 0.7
union
select '40%' as interval, sum(correct)/sum(CASE WHEN label is not null then 1 END) as correct, sum(CASE WHEN label is not null then 1 END) as entries from stocks_reporting 
where certainty not between 0.4 and 0.6
union
select '50%' as interval, sum(correct)/sum(CASE WHEN label is not null then 1 END) as correct, sum(CASE WHEN label is not null then 1 END) as entries from stocks_reporting 
```

```sql
SELECT count(*), avg(percent_change) as avg_loss, count(case when percent_change<0.01 then 1 end) as 1_perc_losses,count(case when percent_change<0.02 then 1 end) as 2_perc_losses,count(case when percent_change<0.05 then 1 end) as 5_perc_losses
FROM stocks_reporting
where correct = 0 and label is not null and certainty not between 0.1 and 0.9
```
