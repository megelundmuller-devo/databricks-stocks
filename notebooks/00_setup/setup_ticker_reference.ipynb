```python
from pyspark.sql.types import StructType, StructField, StringType

schema = StructType([
    StructField("ticker",   StringType(), False),
    StructField("name",     StringType(), False),
    StructField("index",    StringType(), False),
    StructField("country",  StringType(), False),
    StructField("sector",   StringType(), False),
    StructField("currency", StringType(), False),
])

all_tickers = c20_tickers + stoxx50_tickers
rows = [(t["ticker"], t["name"], t["index"], t["country"], t["sector"], t["currency"]) for t in all_tickers]

df = spark.createDataFrame(rows, schema=schema)
df.write.mode("overwrite").format("delta").saveAsTable("stocks.reference.tickers")

print(f"Written {df.count()} tickers to stocks.reference.tickers")
spark.table("stocks.reference.tickers").groupBy("index").count().show()
```
